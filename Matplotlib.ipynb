```python
#legend
x1=[4,5,3,5]
x2=[5,35,24,6]
y1=[9,9,2,4]
y2=[8,3,4,8]
plt.plot(x1,y1,label='Line 1')
plt.plot(x2,y2,label='Line 2')
plt.title("multiple plt")
plt.legend()
```

```python
x1 = [1, 2, 3, 4, 5]
plt.plot(x1, color='red', linestyle='--', marker='o', label='line 1') 
plt.plot(x1, color='blue', linestyle='--', marker='*', label='line 2') 
plt.legend() 
plt.show() 
```

```python
import matplotlib.pyplot as plt
sizes = [15, 30, 45, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D']
plt.pie(sizes, labels=labels)
plt.show()
```
