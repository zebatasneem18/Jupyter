```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Create a simple datasets using dictonary
data={
    
    'Age':[22,23,24,25],
    'salary':[50000,98000,95000,40000],
    'Dep':['HR','IT','Finance','Marketing'],
    'Exp':[2,4,5,8]
}
```

```python
df=pd.DataFrame(data=data)
df
```

```python
x=[22,23,24,25]
y=[50000,98000,95000,40000]
plt.bar(x,y,color='green')
plt.title("Salary")
plt.xlabel("cal")
plt.ylabel("per")
plt.show()
```

```python
x=[22,23,24,25]
y=[50000,98000,95000,40000]
plt.title("Salary")
plt.xlabel("cal")
plt.ylabel("per")
plt.scatter('x','y')
plt.show()
```

```python
x1=[22,23,24,25]
y1=[50000,98000,95000,40000]
x2=[44,55,63,66]
y2=[40000,30000,87000,76000]
plt.plot(x1,y1,label='Line 1')
plt.plot(x2,y2,label='Line 2')
plt.title("multiple plt")
plt.legend()
```

```python
sns.barplot(x='Dep',y='salary',data=data)
plt.show()
```
